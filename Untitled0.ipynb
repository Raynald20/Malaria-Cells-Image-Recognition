# **Data Preparation**
Data preparation analysis consists of :
  1. Set random number generator
  2. Unzip the data file
  3. Dataset labelling
  4. Data file visualization, consists of:

    a. Images visualization
    
    b. Labels countplot visualization





 

```
#Set random number gnerator (RNG)
os.environ['PYTHONHASHSEED'] = '0'
np.random.seed(1000)
rand.seed(1000)
tf.random.set_seed(12345)
session_conf = tf.compat.v1.ConfigProto(intra_op_parallelism_threads = 1,
                              inter_op_parallelism_threads = 1)
seeds = tf.compat.v1.Session(graph = tf.compat.v1.get_default_graph(), config = session_conf)
tf.compat.v1.keras.backend.set_session(seeds)
```

```
#Extract the file from zip file
images = zipfile.ZipFile("/content/cell-images-for-detecting-malaria.zip", 'r')
images.extractall("/content")
images.close()
```

```
#Assign the images into either the infected cells and uninfected cells 

malaria_cells = os.listdir("/content/cell_images/Parasitized")
uninfected_cells = os.listdir("/content/cell_images/Uninfected")
```

```
#Transform all the images into RGB matrices of pixels
#Labelling process is done here
import keras.preprocessing.image as im
picture = []
label = []

for infect in malaria_cells :
  try :
    if infect != 'Thumbs.db':
      image = "/content/cell_images/Parasitized/"+infect
      img_array = im.img_to_array(im.load_img(image), dtype = 'uint8')
      picture.append(np.array(img_array))
      label.append(1)

  except FileNotFoundError:
    continue

for uninfect in uninfected_cells :
  try :
    if uninfect != 'Thumbs.db':
      image = "/content/cell_images/Uninfected/"+uninfect
      img_array = im.img_to_array(im.load_img(image), dtype = 'uint8')
      picture.append(np.array(img_array))
      label.append(0)

  except FileNotFoundError:
    continue
```

```
#Initialization of image data parameters
picture = np.array(picture)
labels = np.array(label)

#Checks the total of rgb images array and the labels of them
np.save('picture', picture)
np.save('labels', labels)

print ('picture : {}, labels : {}'.format(picture.shape, labels.shape))
```

```
#Images visualization before data preprocessing process
plt.figure(num = 1, figsize = (15,15))

for n in range (1,65) :
  val = np.random.randint(0, picture.shape[0], 1)
  plt.subplot(8, 8, n)
  plt.subplots_adjust(wspace = 0.4, hspace = 0.4)
  plt.imshow(picture[val[0]])
  plt.title('{} : {}'.format('Uninfected' if labels[val[0]] == 0 else 'Infected', labels[val[0]]))
  plt.yticks([])
  plt.xticks([])

plt.show()
```

```
#Data amount confirmation
plt.figure(1, figsize = (15,6), dpi = 80)

plt.subplot(1, 3, 1)
sb.countplot (x = labels)
plt.title('Image Data Labels Before Preprocessing')

plt.show()
```
